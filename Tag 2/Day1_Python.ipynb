## 🧩 Operatoren ausprobieren

```python
# Verwende +, -, *, /, % (Rest) und ** (Potenz)
print("3 + 2 =", 3 + 2)
print("10 - 4 =", 10 - 4)
print("6 * 7 =", 6 * 7)
# / (Division) → normale Division, Ergebnis ist immer eine Dezimalzahl (Float)
print("8 / 2 =", 8 / 2)
# // (Ganzzahl-Division) → teilt und rundet das Ergebnis nach unten ab
print("8 // 2 =", 8 // 2)
# % (Modulo) → gibt den Rest der Division zurück
print("10 % 3 =", 10 % 3)
# ** Potenz (Exponent) → also 2 hoch 3
print("2 ** 3 =", 2 ** 3)
```

## 🧩 For-Schleife

```python
# Zahlen von 0 bis 5 ausgeben
for i in range(6):
    print("Zahl:", i)
```

```python
# Zahlen von 1 bis 5 ausgeben
for i in range(1, 6):
    print("Zahl:", i)
```

**Zahlen von 0 bis einschließlich 10 mit Schrittweite 2 ausgeben**

range(start, stop, step) erzeugt eine Zahlenfolge

- start = 0 → Startwert

- stop = 10 → 10 ist nicht enthalten

- step = 2 → Schrittweite 2

```python
# Zahlen von 0 bis einschließlich 10 mit Schrittweite 2 ausgeben
for i in range(0, 10, 2) :
    print("Zahl:", i)
```

```python
farben = ["Rot", "Grün", "Blau"]
for farbe in farben:
    print("Ich mag die Farbe", farbe)
```

## 🧩 While-Schleife

```python
# Countdown von 5 bis 1
x = 5
while x > 0:
    print("Countdown:", x)
    x -= 1 # x = x-1
print("Start!")
```

```python
# Zahlenraten-Spiel
zahl = 3
versuch = 0

while versuch != zahl:
    versuch = int(input("Rate eine Zahl zwischen 1 und 5: "))
    if versuch == zahl:
        print("Richtig!")
    else:
        print("Falsch, versuch es nochmal.")
```

### 🔗 **Daten zusammenführen (Merge & Join in Pandas)**

**Ziel:**

Verstehen, wie man mehrere Tabellen anhand gemeinsamer Spalten zusammenführt (wie in Excel VLOOKUP oder SQL JOIN).

**Wichtige Methode:**

pd.merge(df1, df2, on="gemeinsame_Spalte", **how**="inner")

`how` kann sein:

- `"inner"` – nur gemeinsame Werte (Standard)
- `"left"` – alle aus der linken Tabelle
- `"right"` – alle aus der rechten Tabelle
- `"outer"` – alle Werte, auch wenn sie nicht übereinstimmen

```python
import pandas as pd

# Erste Tabelle: Personen
personen = pd.DataFrame({
    "Name": ["Anna", "Ben", "Clara", "David"],
    "Abteilungs_ID": [1, 2, 2, 3]
})

# Zweite Tabelle: Abteilungen
abteilungen = pd.DataFrame({
    "Abteilungs_ID": [1, 2, 3],
    "Abteilung": ["HR", "Daten", "Marketing"]
})

# Tabellen zusammenführen
merged = pd.merge(personen, abteilungen, on="Abteilungs_ID", how="inner")

print("➡️ Ergebnis:")
print(merged)
```
