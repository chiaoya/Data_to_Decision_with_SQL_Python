## Einrichtung & Importe

Führe diese Zelle zuerst aus. Sie lädt die Bibliotheken, liest die CSV-Datei ein und erstellt daraus eine SQLite-Tabelle namens `budget`.


```python
import sqlite3
from pathlib import Path

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)

DATA_PATH = Path('data_practice_tables/household_budget.csv')
if not DATA_PATH.exists():
    DATA_PATH = Path('Tag 1/data_practice_tables/household_budget.csv')

budget_df = pd.read_csv(DATA_PATH)
budget_df['date'] = pd.to_datetime(budget_df['date'])

conn = sqlite3.connect(':memory:')
budget_df.to_sql('budget', conn, index=False, if_exists='replace')

print('Datensatz geladen:', budget_df.shape[0], 'Zeilen und', budget_df.shape[1], 'Spalten')
budget_df.head()
```

## Aggregationen

Aggregationen verdichten viele Zeilen zu einer Kennzahl. Häufige Funktionen sind:

- `COUNT(*)` zählt Zeilen
- `SUM(amount)` summiert Beträge
- `AVG(amount)` berechnet den Durchschnitt


### Übung 5: Anzahl aller Buchungen

**Aufgabe:** Wie viele Buchungen enthält der Datensatz insgesamt?

**Erwartung:** Der Datensatz hat 60 Zeilen.


```python
# Aufgabe: Zähle alle Buchungen.
query = '''
SELECT COUNT(*) AS anzahl_buchungen
FROM _____;
'''

result = pd.read_sql_query(query, conn)
result
```

### Übung 6: Anzahl der Ausgaben

**Aufgabe:** Wie viele Buchungen sind Ausgaben?

**Hinweis:** Kombiniere `COUNT(*)` mit `WHERE`.


```python
# Aufgabe: Zähle nur Ausgaben.
query = '''
SELECT COUNT(*) AS anzahl_ausgaben
FROM budget
WHERE _____ = 'Expense';
'''

result = pd.read_sql_query(query, conn)
result
```

### Übung 7: Gesamtausgaben berechnen

**Aufgabe:** Wie hoch sind die gesamten Ausgaben?

**Hinweis:** Ausgaben sind negative Beträge. Für eine positive Darstellung kannst du `ABS(SUM(amount))` verwenden.


```python
# Aufgabe: Berechne die gesamten Ausgaben.
query = '''
SELECT SUM(_____) AS gesamtausgaben
FROM budget
WHERE type = 'Expense';
'''

result = pd.read_sql_query(query, conn)
result
```

### Übung 8: Gesamteinnahmen berechnen

**Aufgabe:** Wie hoch sind die gesamten Einnahmen?


```python
# Aufgabe: Berechne die gesamten Einnahmen.
query = '''
SELECT SUM(_____) AS gesamteinnahmen
FROM budget
WHERE type = 'Income';
'''

result = pd.read_sql_query(query, conn)
result
```

### Übung 9: Durchschnittliche Ausgabe

**Aufgabe:** Wie hoch ist eine durchschnittliche Ausgabe?

**Praxisfrage:** Ist der Durchschnitt allein aussagekräftig, oder sollten wir zusätzlich Kategorien betrachten?


```python
# Aufgabe: Berechne den Durchschnitt der Ausgaben.
query = '''
SELECT AVG(_____) AS durchschnittliche_ausgabe
FROM budget
WHERE type = 'Expense';
'''

result = pd.read_sql_query(query, conn)
result
```

## GROUP BY

Mit `GROUP BY` kannst du Kennzahlen pro Gruppe berechnen, zum Beispiel pro Kategorie oder pro Monat.


### Übung 10: Ausgaben pro Kategorie

**Aufgabe:** Wie viel wurde in jeder Kategorie ausgegeben?

**Hinweis:** Nutze `GROUP BY category` und filtere vorher auf `Expense`.


```python
# Aufgabe: Berechne die Ausgaben pro Kategorie.
query = '''
SELECT
    category,
    SUM(amount) AS ausgaben
FROM budget
WHERE type = 'Expense'
GROUP BY _____;
'''

result = pd.read_sql_query(query, conn)
result
```

### Übung 11: Kategorie mit den höchsten Ausgaben

**Aufgabe:** Welche Kategorie verursacht die höchsten Ausgaben?

**Hinweis:** Sortiere die Ergebnisse und begrenze sie mit `LIMIT 1`.


```python
# Aufgabe: Finde die teuerste Kategorie.
query = '''
SELECT
    category,
    ABS(SUM(amount)) AS ausgaben_positiv
FROM budget
WHERE type = 'Expense'
GROUP BY category
ORDER BY _____ DESC
LIMIT _____;
'''

result = pd.read_sql_query(query, conn)
result
```

## Monatsanalyse

Für Monatsanalysen kannst du in SQLite `strftime('%Y-%m', date)` verwenden. Daraus wird zum Beispiel `2024-01`.


### Übung 12: Ausgaben pro Monat

**Aufgabe:** Wie viel wurde in jedem Monat ausgegeben?

**Praxisfrage:** Gibt es einen Monat, der deutlich teurer war als die anderen?


```python
# Aufgabe: Berechne die Ausgaben pro Monat.
query = '''
SELECT
    strftime('%Y-%m', date) AS monat,
    ABS(SUM(amount)) AS ausgaben
FROM budget
WHERE type = 'Expense'
GROUP BY _____
ORDER BY _____;
'''

result = pd.read_sql_query(query, conn)
result
```

### Übung 13: Monat mit den höchsten Ausgaben

**Aufgabe:** Welcher Monat hatte die höchsten Ausgaben?

**Hinweis:** Du kannst die Lösung aus Übung 12 sortieren und mit `LIMIT 1` begrenzen.


```python
# Aufgabe: Finde den teuersten Monat.
query = '''
SELECT
    strftime('%Y-%m', date) AS monat,
    ABS(SUM(amount)) AS ausgaben
FROM budget
WHERE type = 'Expense'
GROUP BY monat
ORDER BY _____ DESC
LIMIT 1;
'''

result = pd.read_sql_query(query, conn)
result
```

### Übung 14: Monatlicher Netto-Betrag

**Aufgabe:** Berechne pro Monat den Netto-Betrag: Einnahmen plus Ausgaben.

**Interpretation:** Ein positiver Wert bedeutet Überschuss, ein negativer Wert bedeutet Verlust.


```python
# Aufgabe: Berechne den Netto-Betrag pro Monat.
query = '''
SELECT
    strftime('%Y-%m', date) AS monat,
    SUM(_____) AS netto_betrag
FROM budget
GROUP BY monat
ORDER BY monat;
'''

result = pd.read_sql_query(query, conn)
result
```

## Bonus: Visualisierung

Wenn du deine SQL-Abfrage gelöst hast, kannst du die Ergebnisse auch als Diagramm darstellen.


```python
# Beispiel: Ausgaben pro Kategorie visualisieren
query = '''
SELECT
    category,
    ABS(SUM(amount)) AS ausgaben
FROM budget
WHERE type = 'Expense'
GROUP BY category
ORDER BY ausgaben DESC;
'''

category_expenses = pd.read_sql_query(query, conn)

sns.barplot(data=category_expenses, x='ausgaben', y='category')
plt.title('Ausgaben pro Kategorie')
plt.xlabel('Ausgaben')
plt.ylabel('Kategorie')
plt.show()
```
