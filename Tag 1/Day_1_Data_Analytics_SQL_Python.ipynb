## 1. Einführung in SQL & Datenbanken

SQL (Structured Query Language) wird verwendet, um relationale Datenbanken zu verwalten und abzufragen.

**Grundstruktur einer Abfrage:**
```sql
SELECT column1, column2
FROM table_name
WHERE condition
ORDER BY column1
LIMIT 10;
```

## 2. Die SELECT-Anweisung

Die SELECT-Anweisung wird verwendet, um Daten aus einer Datenbank abzurufen:

````python
# Verbindung zur Beispieldatenbank herstellen
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Beispieltabelle erstellen
cursor.execute('''
CREATE TABLE kunden (
    id INTEGER PRIMARY KEY,
    name TEXT,
    age INTEGER,
    city TEXT,
    kaufbetrag REAL
)
''')

# Beispieldaten einfügen
data = [
    (1, 'Anna Schmidt', 28, 'Berlin', 150.50),
    (2, 'Max Müller', 35, 'München', 220.00),
    (3, 'Lisa Weber', 32, 'Hamburg', 180.75),
    (4, 'Tom Fischer', 45, 'Berlin', 290.25),
    (5, 'Sarah Klein', 26, 'Köln', 95.00)
]

cursor.executemany('INSERT INTO kunden VALUES (?, ?, ?, ?, ?)', data)
conn.commit()

print('Beispieldatenbank erstellt!')
print('\nDatenbankinhalt:')
df = pd.read_sql_query('SELECT * FROM kunden', conn)
print(df)
````

````python
# Beispiel: Alle Spalten auswählen
query1 = 'SELECT * FROM kunden'
result1 = pd.read_sql_query(query1, conn)
print('Alle Kundinnen und Kunden:')
print(result1)
print(f'Anzahl der Zeilen: {len(result1)}')
````

````python
# Beispiel: Bestimmte Spalten auswählen
query2 = 'SELECT name, age FROM kunden'
result2 = pd.read_sql_query(query2, conn)
print('Namen und Alter:')
print(result2)
````

````python
# Beispiel: WHERE-Klausel - Kundinnen und Kunden aus Berlin filtern
query3 = 'SELECT * FROM kunden WHERE city = "Berlin"'
result3 = pd.read_sql_query(query3, conn)
print('Kundinnen und Kunden aus Berlin:')
print(result3)
````

````python
# Beispiel: Aggregatfunktionen
query4 = 'SELECT COUNT(*) as total_kunden, AVG(age) as durchschnittsalter, SUM(kaufbetrag) as gesamtbetrag FROM kunden'
result4 = pd.read_sql_query(query4, conn)
print('Kundenstatistik:')
print(result4)
````

````python
# Beispiel: ORDER BY und LIMIT
query5 = 'SELECT name, kaufbetrag FROM kunden ORDER BY kaufbetrag DESC LIMIT 3'
result5 = pd.read_sql_query(query5, conn)
print('Top 3 nach Ausgaben:')
print(result5)
````
