## 1. Einführung in SQL & Datenbanken

SQL (Structured Query Language) wird verwendet, um relationale Datenbanken zu verwalten und abzufragen.

**Grundstruktur einer Abfrage:**
```sql
SELECT column1, column2
FROM table_name
WHERE condition
ORDER BY column1
LIMIT 10;
```

````python
# Verbindung zur SQL-Datenbank im Arbeitsspeicher herstellen
conn = sqlite3.connect(':memory:')

# Bestelldaten direkt von GitHub laden
orders_url = "https://raw.githubusercontent.com/chiaoya/Data_to_Decision_with_SQL_Python/refs/heads/main/Tag%201/data_practice_tables/shop_orders.csv"
df = pd.read_csv(orders_url)

# DataFrame als SQL-Tabelle speichern
df.to_sql('bestellungen', conn, if_exists='replace', index=False)

print('Daten aus URL geladen und als SQL-Tabelle bestellungen gespeichert.')
print(df.head())
print(f'Anzahl der Bestellungen: {len(df)}')
````

## 2. SQL-Übungen mit URL-Daten

Wir arbeiten jetzt mit dem DataFrame `df`, der direkt von einer GitHub-URL geladen wurde.
Die Daten wurden zusätzlich als SQL-Tabelle `bestellungen` gespeichert.

**Spalten:**
- `order_id`: Bestellnummer
- `customer_id`: Kundennummer
- `product_id`: Produktnummer
- `quantity`: Bestellmenge

### Übung 1: Alle Bestellungen anzeigen

**Aufgabe:** Schreibe eine SQL-Abfrage, die alle Spalten und alle Zeilen aus `bestellungen` auswählt.

**Hinweis:** Verwende `SELECT * FROM ...`

````python
# Lösung: Alle Bestellungen anzeigen
query = '''
SELECT *
FROM bestellungen;
'''

result = pd.read_sql_query(query, conn)
display(result.head())
````

### Übung 2: Nur bestimmte Spalten auswählen

**Aufgabe:** Zeige nur `order_id`, `customer_id` und `quantity` an.

**Hinweis:** Trenne mehrere Spalten mit Kommas.

````python
# Lösung: order_id, customer_id und quantity auswählen
query = '''
SELECT order_id, customer_id, quantity
FROM bestellungen;
'''

result = pd.read_sql_query(query, conn)
display(result.head())
````

### Übung 3: Bestellungen filtern

**Aufgabe:** Zeige alle Bestellungen, bei denen die Menge `quantity` mindestens 3 ist.

**Hinweis:** Verwende `WHERE quantity >= 3`.

````python
# Lösung: Bestellungen mit quantity >= 3 anzeigen
query = '''
SELECT *
FROM bestellungen
WHERE quantity >= 3;
'''

result = pd.read_sql_query(query, conn)
display(result.head())
````

### Übung 4: Aggregatfunktionen nutzen

**Aufgabe:** Berechne:
- die Anzahl aller Bestellungen
- die gesamte bestellte Menge
- die durchschnittliche Menge pro Bestellung

**Hinweis:** Verwende `COUNT(*)`, `SUM(quantity)` und `AVG(quantity)`.

````python
# Lösung: COUNT, SUM und AVG berechnen
query = '''
SELECT
    COUNT(*) AS anzahl_bestellungen,
    SUM(quantity) AS gesamte_menge,
    AVG(quantity) AS durchschnittliche_menge
FROM bestellungen;
'''

result = pd.read_sql_query(query, conn)
display(result)
````

### Übung 5: Sortieren und begrenzen

**Aufgabe:** Zeige die 5 größten Bestellungen nach `quantity`, absteigend sortiert.

**Hinweis:** Verwende `ORDER BY quantity DESC LIMIT 5`.

````python
# Lösung: Top 5 Bestellungen nach Menge anzeigen
query = '''
SELECT *
FROM bestellungen
ORDER BY quantity DESC
LIMIT 5;
'''

result = pd.read_sql_query(query, conn)
display(result)
````

### Zusatzübung: Bestellungen pro Produkt

**Aufgabe:** Gruppiere die Daten nach `product_id` und berechne pro Produkt die gesamte bestellte Menge.
Sortiere das Ergebnis absteigend nach der gesamten Menge.

**Hinweis:** Verwende `GROUP BY product_id`.

````python
# Lösung: Gesamtmenge pro product_id berechnen
query = '''
SELECT
    product_id,
    SUM(quantity) AS gesamte_menge
FROM bestellungen
GROUP BY product_id
ORDER BY gesamte_menge DESC;
'''

result = pd.read_sql_query(query, conn)
display(result.head())
````

## Abschlussbeispiel: Eine neue Tabelle mit `CREATE TABLE` erstellen

Zum Schluss erstellen wir selbst eine kleine Retail-Sales-Tabelle.
Die Tabelle hat 4 Spalten und 30 Zeilen:

- `sale_id`: eindeutige Verkaufsnummer
- `product_category`: Produktkategorie
- `quantity`: verkaufte Menge
- `revenue`: Umsatz in Euro

Danach fügen wir Beispieldaten ein und lesen die Tabelle wieder mit SQL aus.

````python
# Neue Tabelle retail_sales erstellen
cursor = conn.cursor()

cursor.execute('DROP TABLE IF EXISTS retail_sales')

cursor.execute('''
CREATE TABLE retail_sales (
    sale_id INTEGER PRIMARY KEY,
    product_category TEXT,
    quantity INTEGER,
    revenue REAL
)
''')

retail_sales_data = [
    (1, 'Electronics', 2, 399.98),
    (2, 'Clothing', 5, 124.95),
    (3, 'Home', 3, 89.97),
    (4, 'Sports', 1, 59.99),
    (5, 'Beauty', 4, 79.96),
    (6, 'Electronics', 1, 249.99),
    (7, 'Clothing', 2, 49.98),
    (8, 'Home', 6, 179.94),
    (9, 'Sports', 3, 149.97),
    (10, 'Beauty', 2, 39.98),
    (11, 'Electronics', 3, 599.97),
    (12, 'Clothing', 1, 29.99),
    (13, 'Home', 2, 119.98),
    (14, 'Sports', 4, 199.96),
    (15, 'Beauty', 5, 99.95),
    (16, 'Electronics', 2, 459.98),
    (17, 'Clothing', 6, 179.94),
    (18, 'Home', 1, 34.99),
    (19, 'Sports', 2, 89.98),
    (20, 'Beauty', 3, 54.99),
    (21, 'Electronics', 1, 199.99),
    (22, 'Clothing', 4, 119.96),
    (23, 'Home', 5, 249.95),
    (24, 'Sports', 1, 74.99),
    (25, 'Beauty', 2, 44.98),
    (26, 'Electronics', 4, 799.96),
    (27, 'Clothing', 3, 89.97),
    (28, 'Home', 2, 69.98),
    (29, 'Sports', 5, 249.95),
    (30, 'Beauty', 1, 19.99),
]

cursor.executemany('''
INSERT INTO retail_sales (sale_id, product_category, quantity, revenue)
VALUES (?, ?, ?, ?)
''', retail_sales_data)
conn.commit()

print('Tabelle retail_sales wurde erstellt und mit 30 Zeilen gefüllt.')

query = '''
SELECT *
FROM retail_sales
LIMIT 10;
'''

retail_preview = pd.read_sql_query(query, conn)
display(retail_preview)

summary_query = '''
SELECT
    product_category,
    SUM(quantity) AS total_quantity,
    ROUND(SUM(revenue), 2) AS total_revenue
FROM retail_sales
GROUP BY product_category
ORDER BY total_revenue DESC;
'''

retail_summary = pd.read_sql_query(summary_query, conn)
display(retail_summary)
````
